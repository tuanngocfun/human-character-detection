```python
import os
import shutil

def copy_matching_pairs(image_dir, label_dir, target_image_dir, target_label_dir):
    # Create the target directories if they don't exist
    os.makedirs(target_image_dir, exist_ok=True)
    os.makedirs(target_label_dir, exist_ok=True)

    # Iterate through the label files and look for corresponding image files
    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            image_file = label_file[:-4] + '.jpg'

            # Define source and target paths
            src_image_file = os.path.join(image_dir, image_file)
            src_label_file = os.path.join(label_dir, label_file)
            target_image_file = os.path.join(target_image_dir, image_file)
            target_label_file = os.path.join(target_label_dir, label_file)

            # Check if both the image and label files exist, and if not already in the target directory
            if os.path.exists(src_image_file) and os.path.exists(src_label_file):
                if not os.path.exists(target_image_file):
                    shutil.copy(src_image_file, target_image_file)
                if not os.path.exists(target_label_file):
                    shutil.copy(src_label_file, target_label_file)

# Paths to the old directories (source)
image_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/standard-data/100th/images'
label_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/ultralytics/runs/detect/predict34/labels'

# Paths to the new directories (target)
new_train_image_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/standard-data/activ-checking/complete-first-w-human-real-world-iou-95-70-conf-animation-conf90-iou95/images'
new_train_label_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/standard-data/activ-checking/complete-first-w-human-real-world-iou-95-70-conf-animation-conf90-iou95/labels'

# Copy the corresponding pairs of image and label files
copy_matching_pairs(image_dir, label_dir, new_train_image_dir, new_train_label_dir)

print('Files have been copied successfully.')
```

```python
import os
import shutil

def copy_without_matching_labels(image_dir, label_dir, target_image_dir):
    # Create the target directory if it doesn't exist
    os.makedirs(target_image_dir, exist_ok=True)

    # Iterate through the image files and look for absent corresponding label files
    for image_file in os.listdir(image_dir):
        if image_file.endswith('.jpg'):
            label_file = image_file[:-4] + '.txt'
            
            # Define source and target paths
            src_image_file = os.path.join(image_dir, image_file)
            target_image_file = os.path.join(target_image_dir, image_file)
            
            src_label_file = os.path.join(label_dir, label_file)

            # Check if the image file exists but its label file does not, and if not already in the target directory
            if os.path.exists(src_image_file) and not os.path.exists(src_label_file):
                if not os.path.exists(target_image_file):
                    shutil.copy(src_image_file, target_image_file)

# Paths to the old directories (source)
image_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/standard-data/100th/images'
label_dir = '/home/aivn12gb/yolov8_domain_adaption_evn/ultralytics/runs/detect/predict34/labels'

# Path to the new directory (target)
new_train_image_dir_without_labels = '/home/aivn12gb/yolov8_domain_adaption_evn/standard-data/activ-checking/not-label-with-human-real-world-human-70-conf-iou-95/images'

# Copy the image files that do not have corresponding label files
copy_without_matching_labels(image_dir, label_dir, new_train_image_dir_without_labels)

print('Files without matching labels have been copied successfully.')
```
